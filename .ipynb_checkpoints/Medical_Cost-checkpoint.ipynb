```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import math 
import lightgbm as lgb
import xgboost as xgb

from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor
```

## Feature Engineering

```python
# One-hot encode categorical variables
data_encoded = pd.get_dummies(data, columns=['smoker', 'sex', 'region'], drop_first=False)
categorical_columns = data_encoded.select_dtypes(include='bool').columns
data_encoded[categorical_columns] = data_encoded[categorical_columns].astype(int)
data_encoded
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
correlation_matrix = data_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

There is strong positive correation between someone being a smoker and charges (healthcare costs). There is a strong negative correlation between a non-smoker and having less charges. 

```python
# Selecting the numerical columns 
X = data_encoded.select_dtypes(include = ['float64', 'int64'])

# Adding constant column to account for the intercept
X = add_constant(X)

# Calculate the VIF for each feature
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

No multicollinearity present

## Feature Selection

```python
# Using stepwise regression
X = data_encoded.drop(columns=['charges']) # Dropping dependent variable
y = data_encoded['charges']

# Add a constant to the model (intercept)
X = add_constant(X)

# Stepwise function for backward elimination
def backward_elimination(X, y, significance_level=0.05):
    while True:
        model = sm.OLS(y, X).fit()
        max_p_value = model.pvalues.max()  # Get the feature with the highest p-value
        if max_p_value > significance_level:
            excluded_feature = model.pvalues.idxmax()  # Feature with the highest p-value
            X = X.drop(columns=[excluded_feature])
            print(f"Dropping {excluded_feature} with p-value {max_p_value}")
        else:
            
            break
    return X, model

# Run backward elimination
X_selected, final_model = backward_elimination(X, y)

# Summary of the final model
print(final_model.summary())
```

```python
# Features with the highest predictive power to use in model
X_selected = X_selected.columns
X_selected
```

```python
# Ensure X_selected is a list of selected feature names
selected_columns = list(X_selected)
selected_columns_with_target = selected_columns + ['charges']

# Subset the data_encoded DataFrame using selected columns
numerical_columns = data_encoded[selected_columns_with_target].select_dtypes(include=['float64', 'int64'])

# Calculate skewness for the selected numerical columns
skewness = numerical_columns.skew()
print(skewness)
```

Greater than 1 indicates very skewed. Smoker variables are categorical variables that are encoded so skewness is expected and we can disregard. However, charges is very skewed and needs to be log transformed. 

## Splitting the data for training 

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X[selected_columns], y, test_size=0.2, random_state=42)

# Print the shapes of the resulting sets
print("X_train_selected shape:", X_train.shape)
print("X_test_selected shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Standardize Variables

```python
# Standardize the features before fitting the models
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Lasso Regression

```python
# Fit Lasso Model
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train_scaled, y_train)

# Get the coefficients
coefficients = lasso_model.coef_

# Convert scaled X_train back to DataFrame to get column names
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

# Create a DataFrame for easier viewing
lasso_df = pd.DataFrame({'Feature': X_train_scaled_df.columns, 'Coefficient': coefficients})
lasso_df['Abs_Coefficient'] = np.abs(lasso_df['Coefficient'])  # Add column for absolute value
lasso_df = lasso_df.sort_values(by='Abs_Coefficient', ascending=False)
print(lasso_df)
```

### Evaluating Lasso

```python
# Make predictions on the train set
y_pred_lasso_train = lasso_model.predict(X_train_scaled)

# Calculate the MSE and R^2 for lasso model with selected features for train
mse_lasso_train = mean_squared_error(y_train, y_pred_lasso_train)
r2_lasso_train = r2_score(y_train, y_pred_lasso_train)

print(f"Lasso Train MSE: {mse_lasso_train}")
print(f"Lasso Train R2: {r2_lasso_train}")
```

```python
# Make predictions on the test set
y_pred_lasso_test = lasso_model.predict(X_test_scaled)

# Calculate the MSE and R^2 for lasso model with selected features for test
mse_lasso_test = mean_squared_error(y_test, y_pred_lasso_test)
r2_lasso_test = r2_score(y_test, y_pred_lasso_test)

print(f"Lasso Test MSE: {mse_lasso_test}")
print(f"Lasso Test R2: {r2_lasso_test}")
```

```python
# Calculate the residuals
residuals_train = y_train - y_pred_lasso_train
residuals_test = y_test - y_pred_lasso_test

# Plot the residuals for both training and test sets
plt.scatter(y_pred_lasso_train, residuals_train, color='blue', label='Training Data', alpha=0.5)
plt.scatter(y_pred_lasso_test, residuals_test, color='red', label='Test Data', alpha=0.5)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.legend()
plt.show()
```

The residuals are not randomly scattered around 0, there appears to be clusters of points present. There seems to be a trend in the residuals, which suggests that the model is not capturing the underlying relationships well. Specifically, the pattern seen in clusters could indicate non-linearity in the data that the lasso regression (which is a linear model) is not capturing. Plot is heterscedastic.


### Log transform 'charges' variable

```python
data_encoded['charges_log'] = np.log(data_encoded['charges'] + 1)
```

## Lasso with Log Transform

```python
# Subset data to only include predictors
X = data_encoded[selected_columns]

# Split the data into training and testing sets (with log-transformed 'charges')
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X[selected_columns], data_encoded['charges_log'], test_size=0.2, random_state=42)

# Fit model
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train_log, y_train_log)

# Predict and convert back to the original scale
y_pred_log = lasso_model.predict(X_test_log)
y_pred_original = np.expm1(y_pred_log)  # Inverse transform to original scale

# Evaluate using the original scale of predictions
mse = mean_squared_error(np.expm1(y_test_log), y_pred_original)
r2 = r2_score(np.expm1(y_test_log), y_pred_original)
print(f"Lasso Log-Transformed Test MSE: {mse}")
print(f"Lasso Log-Transformed Test R2: {r2}")
```

```python
# Calculate the residuals
residuals = np.expm1(y_test_log) - y_pred_original  # y_test is still in log scale, so use np.expm1

# Plot residuals vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_pred_original, residuals, color='blue', alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.title('Residual Plot: Residuals vs. Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

The residuals are still heteroscedastic and the model seems to be non-linear so we need to try othe non-linear models.

## Random Forest

```python
# Fit Random Forest Model
# Initialize random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit rf model on train data
rf_model.fit(X_train, y_train) 

# Predict on test data
y_pred_test_rf = rf_model.predict(X_test)

# Predict on train data
y_pred_train_rf = rf_model.predict(X_train)
```

### Evaluating Random Forest

```python
mse_rf_train = mean_squared_error(y_train, y_pred_train_rf)
r2_rf_train = r2_score(y_train, y_pred_train_rf)

print(f"Random Forest MSE Train: {mse_rf_train}")
print(f"Random Fored r^2 Train: {r2_rf_train}")
```

```python
mse_rf_test = mean_squared_error(y_test, y_pred_test_rf)
r2_rf_test = r2_score(y_test, y_pred_test_rf)

print(f"Random Forest MSE Test: {mse_rf_test}")
print(f"Random Fored r^2 Test: {r2_rf_test}")
```

```python
# Feature importance
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```

```python
# Calculate the residuals
train_residuals = y_train - y_pred_train_rf
test_residuals = y_test - y_pred_test_rf

# Plot residuals
plt.scatter(y_pred_train_rf, train_residuals, label='Training Data', color='blue', alpha=0.5)
plt.scatter(y_pred_test_rf, test_residuals, label='Test Data', color='red', alpha=0.5)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot (Test Data)')
plt.legend()
plt.show()
```

The residuals are randomly scattered around 0 and the model seems to capture the non-linearity effectively.

```python
# Select features with importance above a certain threshold (e.g., 0.01)
threshold = 0.01
important_features = feature_importance_df[feature_importance_df['Importance'] > threshold]['Feature']
X_train_important = X_train[important_features]
X_test_important = X_test[important_features]
```

```python
# Fit Random Forest using important features
rf_model_important = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_important.fit(X_train_important, y_train)

# Evaluate the model
y_pred_rf_important = rf_model_important.predict(X_test_important)
mse_rf_important = mean_squared_error(y_test, y_pred_rf_important)
r2_rf_important = r2_score(y_test, y_pred_rf_important)

print(f"Reduced Random Forest MSE Test: {mse_rf_important}")
print(f"Reduced Random Forest R² Test: {r2_rf_important}")
```

## Hyperparamater Tuning and CV for Random Forest

```python
# Define the parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize Grid Search with the reduced feature set
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the grid search using the important features
grid_search.fit(X_train_important, y_train)

# Get the best parameters
best_rf_model = grid_search.best_estimator_
print(f"Best parameters: {grid_search.best_params_}")

# Predict on the test set with the best model
y_pred_best = best_rf_model.predict(X_test_important)

# Evaluate the best model
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print(f"Best Random Forest MSE: {mse_best}")
print(f"Best Random Forest R-squared: {r2_best}")
```

## LightGBM

```python
# Train the LightGBM model using the selected important features
train_data_important = lgb.Dataset(X_train_important, label=y_train)

# Define model parameters for LightGBM
params = {
    'objective': 'regression',     # This is for regression problems
    'metric': 'rmse',              # Root mean squared error for evaluation
    'boosting_type': 'gbdt',       # Gradient boosting decision trees
    'learning_rate': 0.05,         # Step size shrinkage used to prevent overfitting
    'num_leaves': 31,              # Maximum tree leaves for base learners
    'max_depth': -1,               # No limit on tree depth
    'verbose': -1,                 # Suppress LightGBM warnings
    'random_state': 42             # For reproducibility
}

# Train the LightGBM model
num_round = 100  # Number of boosting iterations (can be tuned)
lgb_model_important = lgb.train(params, train_data_important, num_round)

# Predict on the test set using the important features
y_pred_lgb_important = lgb_model_important.predict(X_test_important)

# Evaluate the LightGBM model
mse_lgb_important = mean_squared_error(y_test, y_pred_lgb_important)
r2_lgb_important = r2_score(y_test, y_pred_lgb_important)

print(f"LightGBM (Important Features) Test MSE: {mse_lgb_important}")
print(f"LightGBM (Important Features) Test R²: {r2_lgb_important}")
```

## XGBoost

```python
# Use the important features to train the XGBoost model
# Initialize the XGBoost regressor model with parameters
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',  # Objective for regression
    learning_rate=0.05,            # Step size shrinkage
    n_estimators=100,              # Number of boosting rounds (trees)
    max_depth=6,                   # Maximum depth of trees
    random_state=42,               # For reproducibility
    verbosity=1                    # Verbose output for training process
)

# Fit the model with the important features
xgb_model.fit(X_train_important, y_train)

# Predict on the test set using the important features
y_pred_xgb_important = xgb_model.predict(X_test_important)

# Evaluate the XGBoost model
mse_xgb_important = mean_squared_error(y_test, y_pred_xgb_important)
r2_xgb_important = r2_score(y_test, y_pred_xgb_important)

print(f"XGBoost (Important Features) Test MSE: {mse_xgb_important}")
print(f"XGBoost (Important Features) Test R²: {r2_xgb_important}")
```

## Modeling Final Results

```python
model_names = ['Lasso Regression', 'Random Forest', 'LightGBM', 'XGBoost']
mse_values = [mse_lasso_test, mse_rf_test, mse_lgb_important, mse_xgb_important]
r2_values = [r2_lasso_test, r2_rf_test, r2_lgb_important, r2_xgb_important]

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'Model': model_names,
    'Test MSE': mse_values,
    'Test R-squared': r2_values
})

# Display the results in a neat table
print(results_df)
```

The LightGBM model performed best with a high r^2 value and low test MSE
